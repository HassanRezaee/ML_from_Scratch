```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

ages = np.linspace(25, 35, num=11) # ages
dev_y = np.random.randint(30000, high=75000, size=11)
dev_y = dev_y[np.argsort(dev_y)] # representing salaries

# add a format string [marker][line][color]
fmt1 = "s-g" # dashed green line

plt.plot(ages, dev_y, fmt1, label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

#plt.savefig('salaries.png', dpi=400) # this must come before plt.show()

#plt.show()




py_dev_y = np.random.randint(45000, high=86000, size=11)
py_dev_y = py_dev_y[np.argsort(py_dev_y)] # representing Python developers salaries

# add a format string [marker][line][color]

fmt2 = "o-b"
plt.plot(ages, py_dev_y, fmt2, label="Python Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
#plt.legend(["All Devs","Python Devs"]) # for this you have to know the order in which figures were produced
plt.legend() # this one is better way of adding legend to figure, but have to input the label in plt.plot


plt.savefig('py_dev_salaries.png', dpi=400) # this must come before plt.show()

plt.show()
```

```python
# another way of specifying line color and line types
plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

plt.plot(ages, dev_y, color="g", linestyle="--", marker = "x", label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")


plt.plot(ages, py_dev_y, color="b", linestyle="--", marker = "o", label="Python Devs")
plt.legend()

plt.savefig('py_dev_salaries.png', dpi=400)

plt.show()
```

```python
# adding hex colors

# Hex color palletes are specified as "#fcba03" where consecutive 2 characters are for RGB
plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

plt.plot(ages, dev_y, color="#fcba03", linestyle="--", marker = "x", label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.plot(ages, py_dev_y, color="#14fc03", linestyle="--", marker = "o", label="Python Devs")
plt.legend() 

plt.savefig('py_dev_salaries.png', dpi=400)

plt.show()
```

```python
# adding Java Script developers salaries

plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

js_dev_y = np.random.randint(40000, high=75000, size=11)
js_dev_y = js_dev_y[np.argsort(js_dev_y)] # representing Python developers salaries

plt.plot(ages, dev_y, color="#fcba03", linestyle="--", marker = "x", label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")


plt.plot(ages, py_dev_y, color="#14fc03", linestyle="--", marker = "o", label="Python Devs")
plt.plot(ages, js_dev_y, color="#03dbfc", linestyle="--", marker = "*", label="JS Devs")

plt.legend() 

plt.savefig('py_dev_salaries.png', dpi=400)
plt.show()


```

```python
# Speifying linewidth
plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

plt.plot(ages, dev_y, color="#fcba03", linestyle="--", marker = "x", linewidth=1, label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")


plt.plot(ages, py_dev_y, color="#14fc03", linestyle="--", marker = "o", linewidth=2, label="Python Devs")
plt.plot(ages, js_dev_y, color="#03dbfc", linestyle="--", marker = "*", linewidth=3, label="JS Devs")

plt.legend() 



plt.savefig('js_dev_salaries.png', dpi=400)
plt.show()
```

```python
# Add padding (in case working on a small laptop, figure can have issues on the sides)
# then plt.tight_layout() would help

plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

plt.plot(ages, dev_y, color="#fcba03", linestyle="--", marker = "x", linewidth=1, label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")


plt.plot(ages, py_dev_y, color="#14fc03", linestyle="--", marker = "o", linewidth=2, label="Python Devs")
plt.plot(ages, js_dev_y, color="#03dbfc", linestyle="--", marker = "*", linewidth=3, label="JS Devs")

plt.legend() 
plt.tight_layout()


plt.savefig('js_dev_salaries.png', dpi=400)
plt.show()
```

```python
# Put a grid to the figure

plt.style.use("default")
plt.rcParams['figure.figsize'] = [5, 3.5]

plt.plot(ages, dev_y, color="#fcba03", linestyle="--", marker = "x", linewidth=1, label="All Devs")
plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")


plt.plot(ages, py_dev_y, color="#14fc03", linestyle="--", marker = "o", linewidth=2, label="Python Devs")
plt.plot(ages, js_dev_y, color="#03dbfc", linestyle="--", marker = "*", linewidth=3, label="JS Devs")

plt.legend() 
plt.tight_layout()
plt.grid(True)

plt.savefig('js_dev_salaries.png', dpi=400)
plt.show()
```
