```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
data = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size = 0.2)


# Make blobs
X, y = datasets.make_blobs(n_samples=100, n_features=2, centers=2, cluster_std=1.05)
y = np.where(y==0, -1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
class SVM:
    
    def __init__(self, learning_rate = 0.001, lambda_param=0.01, n_iters = 2000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None
    
    def fit(self, X, y):
        y_ = np.where(y<=0, -1, 1)
        n_samples, n_features = X.shape
        
        # init weights and bias
        self.w = np.zeros(n_features)
        self.b = 0
        
        # Gradient Descent
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    dw = 2 * self.lambda_param * self.w
                    db = 0
                    self.w -= self.lr * dw
                    self.b -= self.lr * db
                else:
                    dw = 2 * self.lambda_param * self.w - np.dot(x_i, y_[idx])
                    db = y_[idx]
                    self.w -= self.lr * dw
                    self.b -= self.lr * db
    
    def predict(self, X):
        linear_model = np.dot(X, self.w) + self.b
        return np.sign(linear_model)
```

```python
clf = SVM()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(f'SVM Classifier Accurate at {100*np.mean(predictions==y_test)} %' )
```

```python
def visualize_svm():
    def get_hyperplane_value(x, w, b, offset):
        return (-w[0] * x +b + offset)/w[1]
    
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    plt.scatter(X[:,0], X[:,1], marker='o', c=y)
    
    x0_1 = np.amin(X[:,0])
    x0_2 = np.amax(X[:,0])
    
    x1_1 = get_hyperplane_value(x0_1, clf.w, clf.b, 0)
    x1_2 = get_hyperplane_value(x0_2, clf.w, clf.b, 0)
    
    x1_1_m = get_hyperplane_value(x0_1, clf.w, clf.b, -1)
    x1_2_m = get_hyperplane_value(x0_2, clf.w, clf.b, -1)
    
    x1_1_p = get_hyperplane_value(x0_1, clf.w, clf.b, 1)
    x1_2_p = get_hyperplane_value(x0_2, clf.w, clf.b, 1)
    
    ax.plot([x0_1, x0_2], [x1_1, x1_2], 'y--')
    ax.plot([x0_1, x0_2], [x1_1_m, x1_2_m], 'k')
    ax.plot([x0_1, x0_2], [x1_1_p, x1_2_p], 'k')
    
    x1_min = np.amin(X[:,1])
    x1_max = np.amax(X[:,1])
    
    plt.show()
    
visualize_svm()
```
