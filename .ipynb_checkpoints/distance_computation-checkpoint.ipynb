```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

'''
Generate some random locations and plot them
'''


np.random.seed(151)
nb_data = random.randint(50,100)
coordinates = np.zeros((nb_data, 2))
coordinates[:,0] = np.random.rand(nb_data)
coordinates[:,1] = np.random.rand(nb_data)
z = np.random.rand(nb_data)*1000

plt.scatter(coordinates[:,0],coordinates[:,1], s=z, c=z, alpha=0.5)
plt.colorbar(label="Z", orientation="vertical")
plt.show()





```

```python
from scipy.spatial.distance import cdist
distances = cdist(coordinates, coordinates)

plt.hist(np.matrix.flatten(distances), bins = 20, color='blue')
plt.title("Histogram of distances computed")
plt.show()
```

```python
# compute kernel densities
kernel_densities = np.zeros((nb_data, nb_data))
kernel_width = np.mean(np.matrix.flatten(distances))
kernel_densities = np.exp(-0.5*(distances/kernel_width)**2)

#import matplotlib.image as mpimg
imgplot = plt.imshow(kernel_densities)
plt.title('Densities computed with Gaussian kernel')
plt.colorbar()
```

```python

class Densities:
        def compute_distances(coordinates, plot_it):
            distances = cdist(coordinates,coordinates)

            if plot_it:
                plt.hist(np.matrix.flatten(distances), bins = 20, color='blue')
                plt.title("Histogram of distances")
                plt.show()

            return distances
        
        def compute_densities(distances, kernel_type, kernel_width, plot_it):
            if kernel_type.lower()=='gaussian':
                kernel_densities = np.exp(-0.5*(distances/kernel_width)**2)
            elif kernel_type.lower()=='triangular':
                u = np.abs(distances/kernel_width)
                u[u<=1] = 0
                kernel_densities = u
            elif kernel_type.lower()=='epanechnikov':
                u = 0.75*( 1 - np.abs(distances/kernel_width)**2 )
                u[u<=1] = 0
                kernel_densities = u
                            
            if plot_it:
                imgplot = plt.imshow(kernel_densities)
                plt.title(f'Densities computed with {kernel_type} kernel')
                plt.colorbar()
                
            return kernel_densities
        
```

```python
distances = Densities.compute_distances(coordinates, False)
kernel_width = np.mean(np.matrix.flatten(distances))
kernel_type = 'gaussian'
densities = Densities.compute_densities(distances, kernel_type, kernel_width, plot_it=True)
```
